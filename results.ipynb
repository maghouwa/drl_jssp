```python
df66 = pd.read_csv("data_6x6.csv").set_index("epoch")
df15 = pd.read_csv("data_15x15.csv").set_index("epoch")
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 12))

df_66_reward = df66["reward"].values
df_66_mean_vloss = df66["mean_vloss"].values

df_15_reward = df15["reward"].values
df_15_mean_vloss = df15["mean_vloss"].values

ax[0,0].plot(df_66_reward)
ax[0,0].set_xlabel("epoch")
ax[0,0].set_ylabel("reward")
ax[0,0].set_title("reward for 6x6 model")

ax[0,1].plot(df_66_mean_vloss)
ax[0,1].set_xlabel("epoch")
ax[0,1].set_ylabel("mean_vloss")
ax[0,1].set_title("mean_vloss for 6x6 model")


ax[1,0].plot(df_15_reward)
ax[1,0].set_xlabel("epoch")
ax[1,0].set_ylabel("reward")
ax[1,0].set_title("reward for 15x15 model")

ax[1,1].plot(df_15_mean_vloss)
ax[1,1].set_xlabel("epoch")
ax[1,1].set_ylabel("mean_vloss")
ax[1,1].set_title("mean_vloss for 15x15 model")
```

```python
import itertools

trained_models = ["6x6", "15x15"]
test_data = ["6x6", "15x15", "20x20", "30x20"]

files = [
    (
        train,
        test,
        f"on_run_results/drlResult_{train}_{test}_Seed200.npy",
        f"on_run_results/drltime_{train}_{test}.txt"
    )
    for (train, test) in itertools.product(trained_models, test_data)
]
files
```

```python
results = []
for (train, test, res_file, time_file) in files:
    res_obj = np.round(np.load(res_file).mean(), 2)
    with open(time_file, "r") as tt:
        res_time = round(float(tt.read().split("/n")[0]), 2)
    results.append([train, test, res_obj, res_time])
results
```

```python
columns = ["model_size", "tst_data_size", "makespan", "instance_timespan"]
result_df = pd.DataFrame(data=results, columns=columns)
result_df
```
